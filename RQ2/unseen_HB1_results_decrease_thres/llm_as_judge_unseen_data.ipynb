```python
from openai import OpenAI
import pandas as pd
import json
import time 
from tqdm import tqdm  
```

```python
client = OpenAI(api_key = '')
```

```python


# 3. Define a function that builds the prompt for the judge LLM.
def build_unseen_data_judge_prompt(question_title, original, generated):
    prompt = f"""
[QUESTION TITLE]
{question_title}
[QUESTION TITLE]

[THE START OF FIRST ANSWER]
{generated}
[THE END OF FIRST ANSWER]

[THE START OF SECOND ANSWER]
{original}
[THE END OF SECOND ANSWER]

"""
    return prompt.strip()
```

```python
SYS_PROMPT = """
We would like to request your feedback on the performance of 2 ANSWERS  in response to the question in [QUESTION TITLE]
Please rate the helpfulness, accuracy, level of details of the responses. 
If you find, there is incompleted code snippet in the response, deduct the overall score.
Each approach receives an overall score on a scale of 1 to 10, where a higher score indicates better overall performance.

Please first output a single line containing only 2 values indicating the scores for FIRST and SECOND ANSWER 
The two scores are separated by a comma(,). 

"""
```

```python
data
```

```python

evaluations = []
scores_list = []

for idx, row in tqdm(data.iterrows(), total=len(data)):
    question_title = row["Question"]
    # accepted_answer = row["Accepted Answer Body"]
    original = row["Accepted_Answer_Body"]
    generated = row["Final_Response"]
    
    # Build the prompt for your judge LLM (gpt-4o-mini)
    prompt_text = build_unseen_data_judge_prompt(
        question_title,
        # accepted_answer,
        original,
        generated
    )
    time.sleep(0.5)
    try:
        
        response = client.chat.completions.create(

        # response = openai.ChatCompletion.create(
            model="gpt-4o",  # Replace with your actual model name
            messages=[
                {"role": "system", "content": SYS_PROMPT},
                {"role": "user", "content": prompt_text}
            ],
            temperature=0.5  # Lower temperature for consistent evaluations
        )
        
        # # Extract the raw text the LLM returned
        # for chunk in response:
        #     print(chunk.choices[0].delta.content or "", end="")
        
        evaluation_text = response.choices[0].message.content

        print(evaluation_text)
        evaluations.append(evaluation_text)
        
    except Exception as e:
        # Handle API/network exceptions, rate limits, etc.
        print(f"OpenAI API error on row {idx}: {e}")
        evaluations.append("Error")
        scores_list.append([None, None, None, None])

```

```python
import numpy as np
from scipy.stats import mannwhitneyu, wilcoxon, ttest_ind, ks_2samp
from cliffs_delta import cliffs_delta

# Perform Mann-Whitney U test
stat, p_value = mannwhitneyu(G, O, alternative='two-sided')

# Output the results
print("Mann-Whitney U statistic:", stat)
print("P-value:", p_value)

delta, interpretation = cliffs_delta(G, O)

# Output results
print("Cliff's Delta:", delta)
print("Effect size interpretation:", interpretation)
```

```python
if len(G) == len(O):
    try:
        stat_wilcoxon, p_wilcoxon = wilcoxon(G, O, alternative='two-sided')
        print(f"Wilcoxon statistic: {stat_wilcoxon}")
        print(f"P-value: {p_wilcoxon}")
    except ValueError as e:
        print(f"Wilcoxon test error: {e}")

# 4. T-Test (for normally distributed, independent samples)
stat_ttest, p_ttest = ttest_ind(G, O, equal_var=False)  # Welch's t-test
print(f"T-test statistic: {stat_ttest}")
print(f"P-value: {p_ttest}")

# 5. Kolmogorov-Smirnov Test (for checking distribution differences)
stat_ks, p_ks = ks_2samp(G, O)
print(f"KS statistic: {stat_ks}")
print(f"P-value: {p_ks}")

# 6. Permutation Test (alternative significance test)
def permutation_test(data1, data2, num_permutations=10000):
    observed_diff = np.abs(np.mean(data1) - np.mean(data2))
    combined = np.concatenate([data1, data2])
    
    count = 0
    for _ in range(num_permutations):
        np.random.shuffle(combined)
        new_data1 = combined[:len(data1)]
        new_data2 = combined[len(data1):]
        new_diff = np.abs(np.mean(new_data1) - np.mean(new_data2))
        if new_diff >= observed_diff:
            count += 1

    p_value = count / num_permutations
    return p_value

p_perm = permutation_test(G, O)
print(f"Permutation test P-value: {p_perm}")
```

## Remove outliars

```python
import numpy as np

def remove_percent_extremes(lst, percent=0.05):
    n = len(lst)
    # Calculate the number of values to remove at each end
    remove_count = int(n * percent)
    # If the list is too short, return the original list
    if remove_count == 0:
        return lst
    # Sort the list and remove the extremes
    sorted_lst = sorted(lst)
    return sorted_lst[remove_count: n - remove_count]


# Remove 5% of the highest and lowest values
G_filtered = remove_percent_extremes(G, percent=0.05)
O_filtered = remove_percent_extremes(O, percent=0.05)

# Calculate new mean and median values
generated_mean = np.mean(G_filtered)
generated_median = np.median(G_filtered)
original_mean = np.mean(O_filtered)
original_median = np.median(O_filtered)


print(f"Generated Mean: {generated_mean}, Generated Median: {generated_median}")
print(f"Original Mean: {original_mean}, Original Median: {original_median}")
```
