```python
import pandas as pd
unseen_data1='../RQ2/unseen_testing/HB1_0.5.csv'
unseen_data2='../RQ2/unseen_testing/HB1_0.6.csv'
unseen_data3='../RQ2/unseen_testing/HB1_0.7.csv'
unseen_data4='../RQ2/unseen_testing/HB1_0.8.csv'
unseen_data5='../RQ2/unseen_testing/HB1_0.9.csv'
RAG4_1 = pd.read_csv(unseen_data1)
RAG4_2 = pd.read_csv(unseen_data2)
RAG4_3 = pd.read_csv(unseen_data3)
RAG4_4 = pd.read_csv(unseen_data4)
RAG4_5 = pd.read_csv(unseen_data5)
```

```python
import re
# Function to check if the CONTEXT is empty
def is_context_empty(cell_text):
    match = re.search(r"### CONTEXT:\s*(\[\s*\])", cell_text)
    return bool(match)  # True if context is exactly "[]", False otherwise
```

```python
RAG4_1['Context_Empty_0.5'] = RAG4_1['Prompt'].apply(is_context_empty)
RAG4_2['Context_Empty_0.6'] = RAG4_2['Prompt'].apply(is_context_empty)
RAG4_3['Context_Empty_0.7'] = RAG4_3['Prompt'].apply(is_context_empty)
RAG4_4['Context_Empty_0.8'] = RAG4_4['Prompt'].apply(is_context_empty)
RAG4_5['Context_Empty_0.9'] = RAG4_5['Prompt'].apply(is_context_empty)
```
